```python
#!python
print("content-type:text/html; charset =UTF-8\n")

#함수 호출


string = '화산 건설 금융 은행 봉사 베트남 북한 축구 월드컵 코스피 코스닥 주식 상승 투데이'

#DB 접속 설정
conn = pymysql.connect(host='ai.cdxxehj8l1sk.ap-northeast-2.rds.amazonaws.com',
                       user='admin', passwd='mypassword' , db='mysql', charset='utf8mb4')
cur=conn.cursor()
cur.execute("USE aiDB")

#페이지 파싱

page = []
y=1
def pages(num):
    if num <200:
        page.append('https://search.naver.com/search.naver?&where=news&query=%EC%9D%B8%EB%8F%84%EB%84%A4%EC%8B%9C%EC%95%84%20-%ED%99%94%EC%82%B0%20-%EA%B1%B4%EC%84%A4%20-%EB%B4%89%EC%82%AC%20-%EB%B2%A0%ED%8A%B8%EB%82%A8%20-%EB%B6%81%ED%95%9C%20-%EC%B6%95%EA%B5%AC%20-%EC%9B%94%EB%93%9C%EC%BB%B5%20-%EC%BD%94%EC%8A%A4%ED%94%BC%20-%EC%BD%94%EC%8A%A4%EB%8B%A5%20-%EC%A3%BC%EC%8B%9D%20-%EC%83%81%EC%8A%B9%20-%ED%88%AC%EB%8D%B0%EC%9D%B4&sm=tab_pge&sort=1&photo=0&field=0&reporter_article=&pd=0&ds=&de=&docid=&nso=so:dd,p:all,a:all&mynews=0&start='+str(num)+'&refresh_start=0')
        pages(num+10)


pages(y)
page.reverse()

page2 = ['https://innipost.com/bbs/board.php?bo_table=business&page=2','https://innipost.com/bbs/board.php?bo_table=business&page=1','https://innipost.com/bbs/board.php?bo_table=politics','https://innipost.com/bbs/board.php?bo_table=world','https://innipost.com/bbs/board.php?bo_table=world&page=2']

#DB에 타이틀 저장

def store(b_title,b_link,b_content,b_id):
    cur.execute(
        "INSERT INTO board_free (b_title,b_link,b_content,b_id) VALUES (\"%s\",\"%s\",\"%s\",\"%s\")",
        (b_title,b_link,b_content,b_id),

    )

    cur.connection.commit()
    global store




#함수 정의


for title in page:
    html = urlopen(title)
    bs = BeautifulSoup(html, "html.parser")
    nameList = bs.find('ul', {'class': 'type01'}).findAll('dt')
    nameList2 = bs.find('ul', {'class': 'type01'}).findAll('a',{'class':'_sp_each_title'})
    nameList3 = bs.find('ul', {'class': 'type01'}).findAll('dd', {'class': ''})
    for name,name2, name3 in zip(nameList,nameList2,nameList3):
        cur.execute("SELECT * FROM board_free WHERE b_title = \"%s\" ", (name.get_text()))
        if cur.rowcount == 0:
            store(name.get_text(),name2.attrs['href'],name3.get_text(),'news')
            


for title in page2:
    html = urlopen(title)
    bs = BeautifulSoup(html, "html.parser")
    nameList = bs.findAll('div', {'class': 'tit'})
    nameList2 = bs.findAll('a',{'class':'bo_tit'})
    nameList3 = bs.findAll('div', {'class': 'txt'})
    for name,name2, name3 in zip(nameList,nameList2,nameList3):
        cur.execute("SELECT * FROM board_free WHERE b_title = \"%s\" ", (name.get_text()))
        if cur.rowcount == 0:
            store(name.get_text(),name2.attrs['href'],name3.get_text()[:100],'indi')
```
