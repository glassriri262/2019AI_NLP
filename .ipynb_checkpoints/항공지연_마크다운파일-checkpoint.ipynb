### 데이터 간략화

```python
import pandas as pd
import numpy as np

from sklearn.impute import SimpleImputer

import warnings

warnings.filterwarnings("ignore")

df = pd.read_csv('AFSNT.csv', engine='python' )
```

```python
df.info()
```

```python
df.drop(['FLT', 'REG', 'SDT_YY', 'SDT_DD', 'CNL','CNR','REG'], axis='columns', inplace=True)
```

```python
df.head()
```

```python
# 전처리를 위해 변수들을 인코딩 해줍니다

from sklearn import preprocessing

def encode_features(dataDF):
    features = ['AOD', 'IRR', 'DLY']
    for feature in features:
        le = preprocessing.LabelEncoder()
        le = le.fit(dataDF[feature])
        dataDF[feature]=le.transform(dataDF[feature])
    return dataDF

df=encode_features(df)
```

- DLY[0] = Delay no
- DLY[1] = Delay yes
- AOD[0] = 도착
- AOD[1] = 출발
- IRR[0] = 정기편
- IRR[1] = 부정기편

- DLT(지연시간) 이상치 값 처리
  - DLT= ATT-STT
  - DLT_R = STT-ATT

```python
# 날짜 계산을 위해 datetime 으로 변환
df_DLY0['STT_N']= pd.to_datetime(df_DLY0['STT'])
df_DLY0['ATT_N']= pd.to_datetime(df_DLY0['ATT'])
```

```python
df_DLY1['STT_N']= pd.to_datetime(df_DLY1['STT'])
df_DLY1['ATT_N']= pd.to_datetime(df_DLY1['ATT'])
```

```python
#DLY=0 인데 DT>30 (실제지연) 인 경우 (*지연은 30분이상이 되는걸 말한다) = 11개

df_DLY0['DT'] = df_DLY0['ATT_N'] - df_DLY0['STT_N']
df_DLY0[(df_DLY0['DT'] > '00:30:00')]
```

```python
df_DLY0 = df_DLY0.loc[(df_DLY0['DT'] <= '00:30:00')]
```

```python
#DLY=1, DT<30(지연이 되지 않음) 
df_DLY1['DT'] = df_DLY1['ATT_N'] - df_DLY1['STT_N']
df_DLY1[(df_DLY1['DT'] < '00:30:00')]
```

```python
df_DLY1 = df_DLY1.loc[(df_DLY1['DT'] >= '00:30:00')]
```

```python
#DLY=0, 데이터중 일찍 출발한 데이터들에서 이상겂들이 많이 발견. 

df_DLY0['DT_D'] = df_DLY0['STT_N'] - df_DLY0['ATT_N']
df_DLY0[(df_DLY0['DT_D'] >= '01:00:00')]
```

```python
df_DLY0 = df_DLY0.loc[(df_DLY0['DT_D'] < '01:00:00')]
```

```python
#전처리 후 데이터 확인
df_DLY1.info()
df_DLY0.info()
```

```python
df_DLY0.drop(['STT_N', 'ATT_N', 'DT', 'DT_D'], axis='columns', inplace=True)
df_DLY1.drop(['STT_N', 'ATT_N', 'DT'], axis='columns', inplace=True)
```

```python
df = pd.concat([df_DLY0, df_DLY1])
```

```python
df.info()
```

```python
# 계절을 기준으로 월별 인코딩 

def badToRight(SDT_MM):
    if SDT_MM in [12, 1, 2]:
        return 4
    elif SDT_MM in [3,4,5]:
        return 1
    elif SDT_MM in [6,7,8]:
        return 2
    elif SDT_MM in [9,10,11]:
        return 3

df['SDT_MM'] = df.SDT_MM.apply(badToRight) 
```

- 12월, 1월, 2월 = 겨울 => 4
- 3월 4월 5월    = 봄   => 1
- 6월 7월 8월    = 여름 => 2
- 9월 10월 11월  = 겨울 => 3

### Resampling

```python
DLY_count = df.DLY.value_counts()
print('DLY 0:', DLY_count[0])
print('DLY 1:', DLY_count[1])
print('Proportion:', round(DLY_count[0] / DLY_count[1], 2), ': 1')
      
      
DLY_count.plot(kind='bar', title='Count (target)')
```

너무 편중된 결과를 보여주기때문에 샘플링이 필요함

```python
count_dly_0, count_dly_1 = df.DLY.value_counts()

df_dly_0 = df[df['DLY'] == 0]
df_dly_1 = df[df['DLY'] == 1]
```

```python
#under-sampling
df_dly_0_under = df_dly_0.sample(count_dly_1)
df_test_under = pd.concat([df_dly_0_under, df_dly_1], axis=0)  

print('Random under-sampling:')
print(df_test_under.DLY.value_counts())

df_test_under.DLY.value_counts().plot(kind='bar', title='Count(target)'); 
```

```python
#over-samling
df_dly_1_over = df_dly_1.sample(count_dly_0, replace=True)  
df_test_over = pd.concat([df_dly_0, df_dly_1_over], axis=0) 

print('Random over-sampling:')
print(df_test_over.DLY.value_counts())

df_test_over.DLY.value_counts().plot(kind='bar', title='Count(target)');  
```
